```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web

sp500 = web.DataReader('^GSPC', data_source='yahoo',start='1/1/2000', end='1/30/2020')
sp500.info()
```

```python
sp500['50d'] = np.round(sp500['Close'].rolling(50).mean(), 2)
sp500['200d'] =np.round(sp500['Close'].rolling(200).mean(), 2)
sp500[['Close','50d','200d']].plot(grid=True, figsize=(15,5))
```
